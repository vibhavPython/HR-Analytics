# 4. Modeling the Data
The best model performance out of the four (Decision Tree Model, AdaBoost Model, Logistic Regression Model, Random Forest Model) was Random Forest!

Note: Base Rate

A Base Rate Model is a model that always selects the target variable's majority class. It's just used for reference to compare how better another model is against it. In this dataset, the majority class that will be predicted will be 0's, which are employees who did not leave the company.
If you recall back to Part 3: Exploring the Data, 24% of the dataset contained 1's (employee who left the company) and the remaining 76% contained 0's (employee who did not leave the company). The Base Rate Model would simply predict every 0's and ignore all the 1's.
Example: The base rate accuracy for this data set, when classifying everything as 0's, would be 76% because 76% of the dataset are labeled as 0's (employees not leaving the company).



```python
# create dummy variable for the 'department' and 'salary' features,since they are categorical

# salary=pd.get_dummies(data_renamed['salary'],drop_first=True,prefix='sal')
# department=pd.get_dummies(data_renamed['department'],drop_first=True,prefix='dep')
# data_reshuffled.drop(['department','salary'],inplace=True,axis=1)
# df=pd.concat([data_reshuffled,salary,department],axis=1)
# df
data_reshuffled['salary']=pd.get_dummies(data_reshuffled['salary'])
data_reshuffled['department']=pd.get_dummies(data_reshuffled['department'])
data_reshuffled
```

```python
from sklearn.model_selection import train_test_split
labels=data_reshuffled['left']
data=data_reshuffled.iloc[:,1:]
X_train,X_test,y_train,y_test= train_test_split(data,labels)
```

```python
#using 5 fold cross validation
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
#linreg=LinearRegression().fit(X_train,y_train)
linRidge=Ridge(alpha=5.0).fit(X_train, y_train)
print('Intercept',linRidge.intercept_)
print('Coefficeint',linRidge.coef_)
print('Training Score:',linRidge.score(X_train,y_train))
print('Test Score:',linRidge.score(X_test,y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
clf=DecisionTreeRegressor().fit(X_train,y_train)
print('Training Score',clf.score(X_train,y_train))
print('Test Score',clf.score(X_test,y_test))
# example_record=[0.38,0.54,5,345,5,0,0,1,1]
# print('predict wether employee will leave or not',clf.predict(example_record))
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression().fit(X_train, y_train)
print('Training score',clf.score(X_train,y_train))
print('Testing score',clf.score(X_test,y_test))
```
